```python
def check_missing(data, percent=False):
    if data.isna().sum().sum()==0:
        return 'No missing values found.'
    else:
        missing= data.isna().sum()[data.isna().sum() > 0].reset_index()
        missing.columns= ['Column','Missing']
        if percent==True:
            missing['percent']= missing.Missing/len(data)
        
        return missing
```

```python
class Plotdata:

    # Intitiate
    def __init__(self, X):
        self.X= X
        self.cat_cols= self.X.select_dtypes('O').columns
        self.num_cols= self.X.select_dtypes(exclude='O').columns
    
    def categorical(self):
        fig, axs= plt.subplots( int(np.ceil(len(self.cat_cols)/3)), 3, figsize= (15,6))
        axs= axs.ravel()
        for ax, col in zip(axs, self.cat_cols):
            sns.countplot(self.X[col], palette='ocean_r', order= self.X[col].value_counts().index, ax=ax)
            ax.set_title(col)
            ax.set_xlabel('')
            ax.set_ylabel('')
        plt.tight_layout()
    
    def numerical(self):
        fig, axs= plt.subplots( int(np.ceil(len(self.num_cols)/3)), 3, figsize= (15,6))
        axs= axs.ravel()
        for ax, col in zip(axs, self.num_cols):
            sns.histplot(self.X[col], palette='ocean_r', ax=ax)
            ax.set_title(col)
            ax.set_xlabel('')
            ax.set_ylabel('')
        plt.tight_layout()
    
    def all(self):
        cols= self.X.columns
        fig, axs= plt.subplots( int(np.ceil(len(cols)/3)), 3, figsize= (15,10))
        axs= axs.ravel()
        for ax, col in zip(axs, cols):
            if self.X[col].dtype=='O':
                sns.countplot(self.X[col], palette='ocean_r', order=self.X[col].value_counts().index, ax= ax)
            else:
                sns.histplot(self.X[col], palette='ocean_r', ax=ax)
            ax.set_xlabel('')
            ax.set_ylabel('')
            ax.set_title(col)
        plt.tight_layout()
```

### Feature engineering and preprocessing

```python
# Function for encoding multiple features
class CustomEncoder:

    def __init__(self, columns):
        self.columns= columns
    
    def fit(self,X,y=None):
        return self
    
    def transform(self, X):
        from sklearn.preprocessing import LabelEncoder
        out= X.copy()
        if self.columns is not None:
            out[self.columns]= out[self.columns].apply(lambda x: LabelEncoder().fit_transform(x))
        else:
            out= out.apply(lambda x: LabelEncoder().fit_transform(x))
        return out
    
    def fit_transform(self, X, y=None):
        out= X.copy()
        return self.fit(out).transform(out)
```

#### Outlier Detection

```python
## Plotting kernel density plots.
fig, axs= plt.subplots(1, 3, figsize= (17,4))
for ax, col in zip(axs,['mileage','mpg','price']):
    sns.kdeplot(np.log(train[col]), ax=ax)
    plt.title(col)
plt.tight_layout()
```

All these 3 features, have a positively skewed distributions(skewed to right), which tells us that outliers are present in these features. we need to get rid of them.

```python
# Function for finding outliers
def outliers(x):
    q3, q1= x.quantile(.75), x.quantile(.25)
    iqr= q3-q1
    li, ri= q1- (1.5*iqr), q3+ (1.5*iqr)
    return (x<li) | (x>ri)
```

```python
# Percent of Outliers in the data:
print('Percent of outliers in price: {:.2%}'.format(outliers(train.price).sum()/len(train)))
print('Percent of outliers in mpg: {:.2%}'.format(outliers(train.mpg).sum()/len(train)))
print('Percent of outliers in mileage: {:.2%}'.format(outliers(train.mileage).sum()/len(train)))
```

### Predictive Modelling

```python
# Importing libraries for predictive analysis 
from sklearn.model_selection import train_test_split, validation_curve, cross_val_score, learning_curve, cross_validate, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, RobustScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error as mse 
from sklearn.metrics import make_scorer, mean_squared_log_error, r2_score
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
import time
```

```python
# Function for calculating RMSE
def rmse(y_true, y_pred):
    from sklearn.metrics import mean_squared_error as mse
    return mse(y_true, y_pred)**0.5
```

```python
# Funtion for getting scores of training multiple ML models:
def multiple_model_scores(estimator_dict, X, y, cv=3):
    multiple_scores=pd.DataFrame()
    for name, model in tqdm(estimator_dict.items()):
        steps= [('le', CustomEncoder(columns=X.columns)), ('sc', RobustScaler()), ('reg',model)]
        pipe= Pipeline(steps= steps)
        scores= cross_validate(estimator=pipe, X= X, y= y, cv=cv, scoring= ('neg_root_mean_squared_error','r2'), return_train_score=True)
        scores= pd.DataFrame(scores)
        scores['model']= name
        multiple_scores= multiple_scores.append(scores, ignore_index=True)
    return multiple_scores.sample(frac=1)
```
